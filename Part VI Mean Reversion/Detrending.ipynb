# Detrending 

$$y_t = y_{0} + \epsilon_{t} + a_0 * t$$

We will use standard Normal Distribution for $\epsilon_{t}$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
n = 500
errors = np.random.normal(0, 1, n)
```

```python
a_0 = 0.5
y0 = 0
y = np.zeros(n)
y[0] = 0.2
y[1]= 0.8
y[2] = -0.6
for i in range(3,n):
    y[i] = y0 + a_0*i + errors[i] + 0.05*y[i-3] + 0.05*y[i-2]+ 0.7*y[i-1]
fig,ax = plt.subplots()
plt.plot(y[:100],label='Time Series')
plt.legend();
```
