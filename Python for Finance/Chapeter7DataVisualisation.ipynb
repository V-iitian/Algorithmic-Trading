## two Dimensional Data set

The following code generates a two-dimensional sample data set as a NumPy ndarray
object of shape 20 × 2 with standard normally distributed pseudo-random numbers.
On this array, the method cumsum() is called to calculate the cumulative sum of the
sample data along axis 0 (i.e., the first dimension)

```python
y = np.random.standard_normal((20,2)).cumsum(axis=0)
```

### Two subplots

The placing of subplots in a matplotlib figure object is accomplished by the use of a
special coordinate system. plt.subplot() takes as arguments three integers for
numrows, numcols, and fignum (either separated by commas or not). numrows speci‐
fies the number of rows, numcols the number of columns, and fignum the number of
the subplot, starting with 1 and ending with numrows * numcols. For example, a fig‐
ure with nine equally sized subplots would have numrows=3, numcols=3, and fig
num=1,2,...,9. The lower-right subplot would have the following “coordinates”:
plt.subplot(3, 3, 9).

```python
plt.figure(figsize=(10,6))
plt.subplot(121) # type of matrix row,cols , figure number
plt.plot(y[:,0],ms=2.0,label='1st')
plt.plot(y[:,0],'ro')
plt.legend(loc='lower left')
plt.subplot(122)
plt.plot(y[:,1],ms=2.0,label='2nd')
plt.plot(y[:,1],'bo')
plt.legend(loc='upper right')
plt.xlabel('index')
plt.ylabel('value of 2nd')
```

Sometimes, it might be necessary or desired to choose two different plot types to vis‐
ualize such data. With the subplot approach one has the freedom to combine arbi‐
trary kinds of plots that matplotlib offers.


```python
y[:,0]
```

### Box plots 

```python
fig,ax = plt.subplots(figsize=(10,6))
plt.boxplot(y)
plt.setp(ax,xticklabels=['1st','2nd'])
plt.xlabel('Data Set')
plt.ylabel('Value')
plt.title('Boxplot')
```

This last example uses the function plt.setp(), which sets properties for a (set of)
plotting instance(s)

The Box (Interquartile Range - IQR): This is the main body of the plot and represents the middle 50% of your data. It shows where the bulk of your typical days fall.

The bottom of the box is the First Quartile (Q1) (25% of your data falls below this line).

The top of the box is the Third Quartile (Q3) (75% of your data falls below this line).

The Median: The solid line drawn inside the box. This is the exact middle value of your dataset (the 50th percentile).

The Whiskers: The lines extending from the top and bottom of the box. They show the typical range of the remaining data outside the middle 50%.

Outliers: Any data points that are unusually extreme (typically calculated mathematically as being more than 1.5 times the size of the box away from the edges). These are plotted as individual, detached dots beyond the whiskers. In trading, these would be your "black swan" events or sudden market shocks.

# Static 3D plotting


There are not too many fields in finance that really benefit from visualization in three
dimensions. However, one application area is volatility surfaces showing implied vol‐
atilities simultaneously for a number of times-to-maturity and strikes of the traded
options used. 

The NumPy np.meshgrid() func‐
tion can generate such a system out of two one-dimensional ndarray object

```python
strike = np.linspace(50,150,24)
```

```python
ttm = np.linspace(0.5,2.5,24)
strike,ttm = np.meshgrid(strike,ttm)
strike[:2].round(1)
```

```python
##  dummy implied volatility values
iv =(strike-100)** 2 / (100*strike) / ttm
```

```python
iv[:5,:3]
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(10,6))
# setting up canvas for 3D plotting 
ax = fig.add_subplot(111, projection='3d')

#Creates the 3d Plot
surf = ax.plot_surface(strike,ttm,iv,
                       rstride=2, # array row stride(step size),
                       cstride=2, # array column stride,
                       cmap=plt.cm.coolwarm, #color map for the surface patches ,
                        linewidth = 0.5,
                        antialiased=True)

ax.set_xlabel('strike')
ax.set_ylabel('time-to-matutiry')
ax.set_zlabel('implied volatility')
# creates the colour bar 
fig.colorbar(surf,shrink=0.5,aspect=5);
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111,projection='3d')

ax.view_init(30,60) # sets the viewing angle 
ax.scatter(strike,ttm,iv,zdir='z',s=25,c='b',marker = '^')

ax.set_xlabel('Strike')
ax.set_ylabel('Time-to-maturity')
ax.set_zlabel('Implied volatility');
```

# Interactive 2D plotting using plotly


In this we will understand and the use of plotly.graph_objects library and will learn its advantages for using it for finance 

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
```

```python
df = pd.read_csv('Oracle.csv')
df['Date']=pd.to_datetime(df['Date'])
# Convert strings with commas to pure numbers, and handle any blank spaces
df['Volume'] = pd.to_numeric(df['Volume'].astype(str).str.replace(',', ''), errors='coerce')
```

```python
df['SMA_20'] = df['Close'].rolling(20).mean()
```

```python
fig = make_subplots(
    rows=2, cols=1,
    shared_xaxes=True,
    vertical_spacing=0.03,
    row_heights=[0.7,0.3] #Gives division of the row (i.e. 70% of the screen to the first and 30% to the other)
)

fig.add_trace(
    go.Candlestick(
        x = df['Date'],
        open = df['Open'],
        high = df['High'],
        low = df['Low'],
        close = df['Close'],
        name='OHLC'
    ),
    row=1,col=1
)

fig.add_trace(
    go.Scatter(
        x=df['Date'],
        y=df['SMA_20'],
        mode='lines',
        line=dict(color='orange',width=2),
        name = '20-Day_sma'
    ),
    row=1,col=1
)

fig.add_trace(
    go.Bar(
        x=df['Date'],
        y=df['Volume'],
        name = 'Volume',
        marker_color = 'blue' # A nice semi transparent blue'
    ),
    row=2 ,col=1
)


fig.update_layout(
    title = 'Strategy Analysis: Price , Sma ,Volume',
    yaxis_title = 'Price',
    yaxis2_title = 'Volume',
    xaxis_rangeslider_visible = False,
    template = 'plotly_white',
    height = 750,
    bargap = 0.2,
    hovermode = 'x unified' # shows a clean crosshair with all data points for a specific day
)

fig.show()
```
