```python
import pandas as pd 
import numpy as np
%matplotlib widget
import matplotlib.pyplot as plt 
```

```python
spot_data = pd.read_csv('BSE Sensex 30 Historical Data (2).csv',index_col='Date',parse_dates=True)
spot_data = spot_data.sort_index()
spot_data = spot_data.dropna()
spot_data = spot_data.drop(columns=['Open','High','Low','Vol.','Change %'])
spot_data =spot_data.replace(',', '', regex=True).astype(float)
d=0 # dividents 
# The risk-free rate in India during 2024 hovered between 6.45% and 7.22%
rf = 6.7/100
spot_data['Fair_Price'] = spot_data['Price']*(1+rf-d)
```

```python
futures_data =  pd.read_csv("BSE Sensex Futures Historical Data (1).csv",index_col='Date',parse_dates=True)
futures_data = futures_data.sort_index()
futures_data = futures_data.drop(columns=['Open','High','Low','Vol.','Change %'])
futures_data = futures_data.replace(',', '', regex=True).astype(float)
futures_data = futures_data.dropna()
futures_data
```

```python
spot_data = spot_data.head(50)
futures_data = futures_data.head(50)
plt.plot(spot_data['Price'],'b-',label='Spot_data')
plt.plot(futures_data['Price'],'ro',label = 'Futures_data')
plt.plot(futures_data['Price'],'g-',label = 'Fair_Price_Futures')
plt.legend();


# its pretty much evident that the futures price was exactly near the Fair Price 
```

```python
spot_data = spot_data.iloc[50:100]
futures_data = futures_data.iloc[50:100]
plt.plot(spot_data['Price'],'b-',label='Spot_data')
plt.plot(futures_data['Price'],'ro',label = 'Futures_data')
plt.plot(futures_data['Price'],'g-',label = 'Fair_Price_Futures')
plt.legend();


# its pretty much evident that the futures price was exactly near the Fair Price 
```
