## Methods 

Methods are function that belong to objects 

```python
class Student:
    def __init__(self,name,marks):
        self.name = name 
        self.marks = marks 
    # methods 
    # self is always passed 
    def hello(self,name):
        print('hello',name,"from",self.name)

    def get_marks(self):
        print(self.marks)

        
```

```python
s1 = Student('Vipul',0)
s1.get_marks()
```

## Practice Question 1 

Create student class that takes name and marks of 3 subjects as arguments in constructors, Then create a method to print message 

```python
class Student:
    def __init__(self,name,marks1,marks2,marks3):
        self.name = name
        self.Physics_marks = marks1
        self.Maths_marks = marks2
        self.Chemistry_marks = marks3

    def get_average(self):
        return (self.Physics_marks+self.Maths_marks+self.Chemistry_marks)/3
```

```python
s1 = Student('Vipul',91,92,34)
s1.get_average()
```

## Static Methods 

Methods that do not use the self parameter(work at class level)

**Decorator allows us to wrap another function in order to extend the behaviour of the wrapped function, without permanently modifying it 

```python
class Student:
    def __init__(self,name,marks1,marks2,marks3):
        self.name = name
        self.Physics_marks = marks1
        self.Maths_marks = marks2
        self.Chemistry_marks = marks3

    def get_average(self):
        return (self.Physics_marks+self.Maths_marks+self.Chemistry_marks)/3
    # declaration to create static methods 
    @staticmethod  # this is a decorator 
    def hello():
        print('hello')
```

## Abstraction 


Hiding the implementation details of a class and only showing the essential feature to the user 

```python
class Car:
    def __init__(self):
        self.acc = False
        self.brk = False
        self.clutch = False

    def start(self):
        self.acc = True
        self.clutch = True
        print('Car started')


c1 = Car()
c1.start()
# you hid the internal details regarding brake ,accelerator ,etc but only showed the necessary details 
```

## Question 2

Create Account class with 2 attributes - balance and account no.

Create methods for debit, credit and printing balance 

```python
class Account:
    def __init__(self):
        self.Amount = 0
    
    def __init__(self,initial_deposit):
        self.Amount = initial_deposit

    def Debit(self,Amount):
        self.Amount-=Amount
        print(Amount,"has been debited from your bank account ")

    def Credit(self,Amount):
        self.Amount+=Amount
        print(Amount,"has been credited to your bank account ")

    def Balance(self):
        print(f"Your current bank balance is {self.Amount} ")


acc1 = Account(10000)
```

```python
acc1.Debit(1000)
```

```python
acc1.Credit(100)
```

```python
acc1.Balance()
```
